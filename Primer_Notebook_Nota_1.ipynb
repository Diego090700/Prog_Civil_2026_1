# 1. Problema de Ingeniería

Un **ingeniero agrícola** necesita calcular el **volumen de agua almacenado en un tanque rectangular** para riego.

La fórmula del volumen es:

[
V = largo \times ancho \times altura
]

donde:

* (largo) = longitud del tanque (m)
* (ancho) = ancho del tanque (m)
* (altura) = altura del agua (m)
* (V) = volumen del agua (m³)

---

# 2. Definición del algoritmo

### Entradas (Inputs)

Datos que el usuario debe ingresar:

* largo del tanque
* ancho del tanque
* altura del agua

---

### Proceso (Processing)

1. Leer los valores de largo, ancho y altura.
2. Multiplicar las tres variables.
3. Guardar el resultado en la variable **volumen**.

---

### Salida (Output)

Mostrar en pantalla:

* El **volumen de agua almacenado en el tanque** en metros cúbicos.

---

# 3. Pseudocódigo

Antes de programar en Python, escribimos la lógica.

```
Inicio

Leer largo
Leer ancho
Leer altura

volumen = largo * ancho * altura

Mostrar "El volumen del tanque es:", volumen, "m3"

Fin
```

---

# 4. Código en Python

```python
# -----------------------------------------
# Programa: Cálculo del volumen de un tanque
# Aplicación: Ingeniería agrícola / civil
# -----------------------------------------

# Entrada de datos
largo = float(input("Ingrese el largo del tanque (m): "))
ancho = float(input("Ingrese el ancho del tanque (m): "))
altura = float(input("Ingrese la altura del agua (m): "))

# Proceso
volumen = largo * ancho * altura

# Salida
print("El volumen de agua almacenado es:", volumen, "m^3")
```

---

# 5. Ejemplo de ejecución

Entrada del usuario:

```
Largo = 5 m
Ancho = 3 m
Altura = 2 m
```

Resultado:

```
El volumen de agua almacenado es: 30 m^3
```
